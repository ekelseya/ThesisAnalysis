```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import researchpy as rp
from scipy import stats

# Read in csv
df = pd.read_csv('fittedFailureRate.csv', encoding='unicode_escape')
df.head()
```

```python
model = ols("""FR_fitted ~ C(Type)""", data=df).fit()
sm.stats.anova_lm(model, typ=1)
```

```python
model = ols("""FR_fitted ~ C(Region)""", data=df).fit()
sm.stats.anova_lm(model, typ=1)
```

```python
model = ols("""FR_fitted ~ C(Page_Type)""", data=df).fit()
sm.stats.anova_lm(model, typ=1)
```

```python
model = ols("""FR_fitted ~ C(Page_Type) + C(Region) + C(Type)""", data=df).fit()
sm.stats.anova_lm(model, typ=1)
```

```python
model = ols("""FR_fitted ~ C(Type) + C(Region) + C(Page_Type)""", data=df).fit()
sm.stats.anova_lm(model, typ=1)
```
